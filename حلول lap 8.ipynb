```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('company_sales_data.csv')
```

```python
plt.plot(df,linestyle= '--' , color='red', linewidth = 1 ,marker = 'o', mfc='k' ,label ="company profit per month",)
plt.xlabel("Month Number")
plt.ylabel("profit in dollar")
plt.ylim(100000)
plt.yticks(range(100000,600000,100000))
plt.xticks(range(1,13))
plt.show()
```

```python
plt.plot(df['month_number'],df['facecream'], linestyle='-', label="Face cream Sales Data", marker='o' )
plt.plot(df['month_number'],df['facewash'], linestyle='-', label="Face Wash Sales Data", marker='o' )
plt.plot(df['month_number'],df['toothpaste'], linestyle='-', label="ToothPaste Sales Data", marker='o' )
plt.plot(df['month_number'],df['bathingsoap'], linestyle='-', label="Bathingshop Sales Data", marker='o' )
plt.plot(df['month_number'],df['moisturizer'], linestyle='-', label="Moisturizer Sales Data", marker='o' )
plt.plot(df['month_number'],df['shampoo'], linestyle='-', label="Shampoo Sales Data", marker='o' )
plt.xlabel("Months Number")
plt.ylabel("Sales units in number")
plt.legend(loc=2)
plt.ylim(1000)
plt.xticks(range(1,13))
plt.show()
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(2, 2, 1)
plt.plot(np.arange(6), 'k--*')
ax2 = fig.add_subplot(2, 2, 2)
plt.plot(2*np.arange(6), 'b-o')
ax3 = fig.add_subplot(2, 2, 3)
plt.plot((np.arange(6))**2,linestyle= '--' , color='g', linewidth = 1 ,marker = '^', mfc='k' )
ax4 = fig.add_subplot(2, 2, 4)
plt.plot((np.arange(6))**3, linestyle= '-.' , color='yellow', linewidth = 2 ,marker = 's', mfc='r' )
```
