```python
import pandas as pd
a = {'names': ['ahmed', 'mohammed', 'anas', 'foad', 'loay', 'gamal', 'ahmed'],
'ids': ['987643776', '846993756', '778354653', '892347346', '312278543', '773536473', '783526234'],
'ages': [62, 27, 20, 50, 43, 22, 34],
 'number of children' : [4, 2, 1, 4, 5, 1, 3],
 'years of service':[30, 2, 1, 22 , 15, 1, 12]}
data = pd.DataFrame(a)
data = data.sort_values(by=['names', 'ids'])
data = data[data['ages'] > 30]  # filter by age
data = data[data['ages']-data['years of service']>24]  # filter by years of service
data
```

```python
market = pd.DataFrame({'names': ['sugar', 'milk', 'soda', 'chocolate', 'meat', 'yogurt', 'cake'],
 'price': [15, 32, 3, 6, 35, 4, 12],
 'quantity': [150, 243, 300, 60, 30, 456, 470],
 'sold quantity per month' : [120, 150, 313, 58, 8, 359, 60]})
market.set_index('names',inplace=True)
market
```

```python
market.loc[market["sold quantity per month"]>market["quantity"] ,:] = market["quantity"]
market= market.sort_values(by='sold quantity per month',ascending=False)
market["rate"]=market["sold quantity per month"]/market["quantity"]
market
```

```python
soldData = market["quantity"]-market["sold quantity per month"]
soldData = soldData.sort_values()
soldData
```

```python
market =  market.sort_values(by='rate',ascending=False)
print("more" , market.head(3).index.values)
print("must stop buy", market[market["rate"]<0.3].index.values)
market
```
